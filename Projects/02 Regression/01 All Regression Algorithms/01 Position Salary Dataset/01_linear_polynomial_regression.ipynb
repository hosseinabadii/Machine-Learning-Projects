# Polynomial Regression

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('data/Position_Salaries.csv')
X = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, -1].values
```

```python
X
```

```python
dataset
```

## Training the Linear Regression model

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

## Training the Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=4)
X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

```python
X_poly
```

## Predicting a new result with Linear Regression

```python
pos = np.array([[6.5]])
lin_reg.predict(pos)
```

## Predicting a new result with Polynomial Regression

```python
pos = np.array([[6.5]])
lin_reg_2.predict(poly_reg.fit_transform(pos))
```

## Visualising the Linear Regression results

```python
plt.scatter(X, y, color="red", label="Real")
plt.plot(X, lin_reg.predict(X), color="blue", label="Predicted")
plt.title("Linear Regression")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.legend()
plt.show()
```

## Visualising the Polynomial Regression results

```python
plt.scatter(X, y, color="red", label="Real")
plt.plot(X, lin_reg_2.predict(X_poly), color="blue", label="Predicted")
plt.title("Polynomial Regression")
plt.xlabel("Position level")
plt.ylabel("Salary")
plt.legend()
plt.show()
```
