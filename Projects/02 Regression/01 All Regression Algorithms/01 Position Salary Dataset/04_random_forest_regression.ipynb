# Random Forest Regression

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('data/Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```python
dataset
```

## Training the Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=10, random_state=0)
regressor.fit(X, y)
```

## Predicting a new result

```python
y_pred = regressor.predict([[6.5]])
print(f"Prediction with Random Forest: {y_pred[0]}")
```

## Visualising the Random Forest Regression

```python
X_grid = np.arange(X.min(), X.max(), 0.01)
X_grid = X_grid.reshape(-1, 1)
plt.scatter(X, y, color="red", label="Real")
plt.plot(X_grid, regressor.predict(X_grid), color="blue", label="Predicted")
plt.title("Random Forest Regression")
plt.xlabel("Position level")
plt.ylabel("Salary")
plt.legend()
plt.show()
```
