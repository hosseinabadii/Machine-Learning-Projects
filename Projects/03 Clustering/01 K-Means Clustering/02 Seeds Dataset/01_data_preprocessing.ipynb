```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load Dataset

```python
pd.read_csv("data/seeds_dataset.txt").head()
# We have to specify the seprator as sep="\t+" or sep="\s+"
```

```python
# The seprator is all white spaces.
pd.read_csv("data/seeds_dataset.txt", sep="\s+").head()
# We have to specify the name of columns
```

```python
cols = [
    "area", "perimeter", "compactness", "length",
    "width", "asymmetry", "groove", "class",
]
df = pd.read_csv("data/seeds_dataset.txt", sep="\s+", names=cols)
df.head()
# The final dateframe is ready
```

## Data Visualization

```python
plt.figure(figsize=(10, 20))
n = 0
for i in range(len(cols) -1):
    for j in range(i+1, len(cols)-1):
        n += 1
        plt.subplot(7, 3, n)
        sns.scatterplot(
            data=df, x=cols[i], y=cols[j],
            hue="class",
            style="class",
            palette=["red", "green", "blue"],
            legend=False
        )
plt.tight_layout(pad=2)
plt.show()
```
